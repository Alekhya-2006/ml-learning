## Calculus

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot y = √x

# define x values (x >= 0 for square root)
x = np.linspace(0, 100, 100_000)
y = np.sqrt(x)

# Plot
plt.plot(x, y, label = "y = √x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = √x")
plt.legend()
plt.grid(True)

plt.show()
```
