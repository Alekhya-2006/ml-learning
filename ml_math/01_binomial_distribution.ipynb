## Binomial Distribution
- Used to model the number of successes in N trials.
- Formuls: P(X = k) = C(n, k) * P^k * (1 - P)^(n - k)
  - where:
    - n  = number of trials
    - k = number of successes
    - P = probability of success
    - C(n, k) is number of ways to choose k successes out of n (Combinations)

```python
from scipy.stats import binom

n = 5
k = 3
p = 0.5

prob = binom.pmf(k, n, p)
print(prob)
```

An experiment follows a binomial distribution if:
1. There are fixed number of trials (n)
2. Each trial has only two outcomes
3. Probability of success p is same every time
4. Trials are independent
