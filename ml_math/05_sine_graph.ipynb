```python
import numpy as np
import matplotlib.pyplot as plt
```

### Sine Function

```python
# Create x values from -2π to 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)

# Highlight x = 0 and y = 0 lines
plt.axvline(0, color = "black", linewidth = 1.5, linestyle = "--", label = "y = 0")
plt.axhline(0, color = "black", linewidth = 1.5, linestyle = "--", label = "x = 0")

# Plot the Graph
plt.plot(x, y, label = "y = sin(x)", color = "green")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of y = sin(x)")
plt.legend()
plt.grid(True)

plt.show()
```
